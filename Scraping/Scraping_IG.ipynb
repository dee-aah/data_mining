```python
from snscrape.modules.twitter import TwitterSearchScraper
import pandas as pd

data = []
query = "Banjir Sumatra:id"

for i, tweet in enumerate(TwitterSearchScraper(query).get_items()):
    data.append({
        "username": tweet.user.username,
        "text": tweet.content,
        "date": tweet.date,
        "like": tweet.likeCount,
        "retweet": tweet.retweetCount
    })
    if i >= 200:
        break

df = pd.DataFrame(data)
df
```

```python
df.to_csv("komentar_instagram.csv", index=False)
print("File komentar_instagram.csv berhasil disimpan")
```
