```python
import pandas as pd
import math
import os
import matplotlib.pyplot as plt
import seaborn as sns

def hitung_entropi(kolom_target):
    """Menghitung Entropi."""
    nilai_counts = kolom_target.value_counts()
    total_data = len(kolom_target)
    entropi = 0
    
    print(f"--- Detail Perhitungan untuk '{kolom_target.name}' ---")
    for nilai, count in nilai_counts.items():
        p = count / total_data
        if p > 0:
            e_part = -p * math.log2(p)
            entropi += e_part
            print(f"Nilai: {nilai} | Jumlah: {count} | Probabilitas: {p:.2f}")
            
    return entropi
```

```python
nama_file = 'car_evaluation.csv' 

if os.path.exists(nama_file):
    df = pd.read_csv(nama_file)
    print("Data berhasil dimuat!")

    display(df.head(21)) 
else:
    print("File tidak ditemukan!")
```

```python
sns.scatterplot(data=df, x=df["Buying"], y=df["Maint"], hue=df["Class"])
plt.show()
```

```python
target_kolom = 'Class' 

if target_kolom in df.columns:
    hasil = hitung_entropi(df[target_kolom])
    print("=" * 40)
    print(f"TOTAL ENTROPI : {hasil:.4f}")
else:
    print(f"Kolom {target_kolom} Tidak Ditemukan.")
```
